```python
import streamlit as st
import pandas as pd
import plotly.express as px

@st.cache_data
def load_data():
    df = pd.read_csv('./data/Coffee Shop Sales.csv', index_col='transaction_date', parse_dates=True)# Adjust column name
    return df

df = load_data()

# Sidebar filter for date range
start_date = st.sidebar.date_input("Start Date", df["transaction_date"].min())
end_date = st.sidebar.date_input("End Date", df["transaction_date"].max())

filtered_df = df[(df["transaction_date"] >= pd.to_datetime(start_date)) & 
                 (df["transaction_date"] <= pd.to_datetime(end_date))]

# Time Series Plot
fig = px.line(filtered_df, x="transaction_date", y="transaction_qty	", title="Time Series Visualization")
st.plotly_chart(fig)
```
